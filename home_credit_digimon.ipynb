## Normalisasi

```python
df = pd.get_dummies(df1)
```

```python
X = df.drop(['TARGET'],1)
y = df['TARGET']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
X_transform = pd.DataFrame(X_transform,columns = X.columns)
```

## Hasil normalisasi data agar antar feature memiliki range yang tidak jauh beda

```python
X_transform
```

# Train test split Validation

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.3,random_state = 123)
```

## Modeling dan evaluasi model

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

## KNN

```python
def evaluasi_model(model,X_test,y_test):
    from sklearn.metrics import accuracy_score
    y_pred = model.predict(X_test)
    return accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
evaluasi_model(knn,X_test,y_test)
```

```python
pred1 = knn.predict(X_test)
print(confusion_matrix(y_test, pred1))
print(classification_report(y_test, pred1))
```

## Decission Tree

```python
from sklearn import tree

decision_tree = tree.DecisionTreeClassifier()
model_pertama = decision_tree.fit(X_train, y_train)
model_pertama.score(X_test, y_test)
```

```python
pred2 = decision_tree.predict(X_test)
print(confusion_matrix(y_test, pred2))
print(classification_report(y_test, pred2))
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)

# mengaplikasikan PCA pada dataset
pca_attributes = pca.fit_transform(X_train)

# melihat variance dari setiap atribut
pca.explained_variance_ratio_
```

```python
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

## KNN

```python
knn.fit(X_train_pca,y_train)
evaluasi_model(knn,X_test_pca,y_test)
```

```python
pred11 = knn.predict(X_test_pca)
print(confusion_matrix(y_test, pred11))
print(classification_report(y_test, pred11))
```

## Decision Tree

```python
model2 = decision_tree.fit(X_train_pca, y_train)
model2.score(X_test_pca, y_test)
```

```python
pred21 = decision_tree.predict(X_test_pca)
print(confusion_matrix(y_test, pred21))
print(classification_report(y_test, pred21))
```
